Example: A Growth Model
==

```julia
B=10; M=5; alpha=0.5; beta=0.9
u(c) = c^alpha

n=B+M+1
m=M+1

R = Array(Float64,n,m)
Q = zeros(Float64,n,m,n)

[R[s+1,a+1] = a<=s ? u(s-a) : -Inf for a=0:M, s=0:(B+M)]
[Q[:,a+1,(a:(a+B))+1] = 1/(B+1) for a=0:M];
```

```julia
using QuantEcon
ddp = DiscreteDP(R,Q,beta);
```

```julia
results = solve(ddp,PFI);
```

```julia
fieldnames(results)
```

```julia
results.v
```

```julia
using PyPlot
bar(0:(B+M),mc_compute_stationary(results.mc));
```

```julia
ddp = DiscreteDP(R,Q,0.99)
results = solve(ddp,PFI)
bar(0:(B+M),mc_compute_stationary(results.mc));
```

```julia
s_indices = Int64[]
a_indices = Int64[]
Q = Array(Float64,0,n)
R = Float64[]
b = 1.0 / (B + 1)

for s in 0:(M+B)
    for a in 0:min(M, s)
        s_indices = [s_indices; s+1]
        a_indices = [a_indices; a+1]
        q = zeros(Float64,1,n)
        q[(a+1):((a + B)+1)] = b 
        Q = [Q; q]
        R = [R; u(s-a)]
    end
end
```

```julia
ddp = DiscreteDP(R,Q,beta,s_indices,a_indices);
results = solve(ddp,PFI)
bar(0:(B+M),mc_compute_stationary(results.mc));
```
